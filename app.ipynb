```python
import streamlit as st
import numpy as np
import tensorflow as tf
import joblib
```

```python
model=tf.keras.models.load_model("student_model_x.h5")
```

```python
scalar=joblib.load("scaler_student_performance.pkl")

```

```python
#user inputs
''' Index(['Hours_Studied', 'Attendance', 'Sleep_Hours',
       'Parental_Education_Level', 'Previous_Scores', 'Gender',
       'Motivation_Level', 'Extracurricular_Activities', 'Exam_Score'],
      dtype='object') '''
```

```python
## User Inputs
Hours_Studied=st.slider("hours Studied",0,12,6)
Attendance=st.slider("Attendance",10,100,60)
Sleep_Hours=st.slider("Sleep Hours",10,12,8)
Parental_Education_Level=st.slider("Parent Education Level",0,1,2)
Previous_Scores=st.slider("Previous Score",0,100,70)
Gender=st.selectbox("Gender",[0,1])
Motivation_Level=st.selectbox("Motivation Level",[0,1,2])
Extracurricular_Activities=st.selectbox("Extra Curricular",[0,1])
```

```python
import numpy as np

input_data = np.array([[
    Hours_Studied,
    Attendance,
    Sleep_Hours,
    Parental_Education_Level,
    Previous_Scores,
    Gender,
    Motivation_Level,
    Extracurricular_Activities
]], dtype=float)

```

```python
input_scaled=scalar.transform(input_data)
```

```python
## 
```

```python
if st.button("Pridict Score"):
    prediction=model.predict(input_scaled)
    st.success(f'Pridected Exam Score: {Prediction[0][0]:2f}')
```
